```python
#import all the neccessary modules
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read the data from csv file
data = pd.read_csv(r"D:\weatherHistory.csv")
```

```python
#take a look at first five rows using head
data.head()
```

```python
data.info()
```

```python
#converting formatted date to datetime format
data['Formatted Date']=pd.to_datetime(data['Formatted Date'],utc=True)
```

```python
data.head()
```

#  precipitation in weather

Now , we are going to analyze the type of  precipitaion like snow,rain , icecrystals , etc.,

```python
sns.countplot(x='Precip Type' ,data = data  )
```

From the count plot we can understand that there is rain most of the time.

# Analyzing of  Humidity during snow and rain

Now , we will analyse at what level of humidity snow fall occurs

```python
sns.boxplot(x='Precip Type',y='Humidity',data = data)
```

It clearly depicts that when the humidity is above 0.8 snowfall occurs and when it is between 0.6 and 0.85 rain occurs.

# Analyzing of temperature during rain and snow

```python
sns.boxplot(x='Precip Type',y='Temperature (C)',data = data)
```

This graph shows that , whenever the temperature goes below 0 degree snowfall occurs and when temperature is between 8(on an average) and 20,there is rain

# Analyzing wind speed during rain and snow

```python
sns.boxplot(x='Precip Type',y='Wind Speed (km/h)',data = data)
```

The wind speed doesn't crossed 9km/h during snow. 

# Analyzing visibility during rain and snow

```python
sns.boxplot(x='Precip Type',y='Visibility (km)',data = data)
```

It is quite obvious that during snow the visibility is low as compared to that of rain

# Analyzing Pressure during rain and snow

```python
sns.barplot(x='Precip Type',y='Pressure (millibars)',data = data)
```

There is no difference in atmosphiric pressure during rain and snow

# Analyzing Wind Bearing(degrees) during rain and snow

```python
sns.barplot(x='Precip Type',y='Wind Bearing (degrees)',data=data)
```

Wind bearing doesn't differ during rain and snow

## Droping unwanted feilds

```python
#set index as formatted date
data = data.set_index('Formatted Date')
```

```python
data.head()
```

```python
#dropping unwanted columns
```

```python
data = data.drop(['Daily Summary','Wind Bearing (degrees)','Summary','Precip Type','Temperature (C)','Wind Speed (km/h)','Visibility (km)','Pressure (millibars)'], axis = 1)
```

```python
data.head()
```

```python
#resampling the data
data = data[['Apparent Temperature (C)','Humidity']].resample('MS').mean()
```

```python
data.head()
```

# Apparent temparature v/s humidity

```python
plt.figure(figsize = (10,6))
plt.plot(data['Humidity'], label = 'Humidity', color = 'violet');
plt.plot(data['Apparent Temperature (C)'], label = 'Apparent temperature',color = 'green');
plt.title('Apparent temparature v/s Humidity')
plt.legend()
```

# Humidity vs Temperature during april

```python
april = data[data.index.month==4]
plt.figure(figsize=(10,5))
plt.plot(april.loc['2006-04-01':'2016-04-01', 'Apparent Temperature (C)'], marker='o', linestyle='-',label='Apparent Temperature (C)',color = 'green')
plt.plot(april.loc['2006-04-01':'2016-04-01', 'Humidity'], marker='o', linestyle=':',label='Humidity',color = 'violet')
plt.title("humidity vs temperature")
plt.legend()
plt.xlabel("during the month of april")
plt.ylabel("values")
```

```python
data.reset_index(inplace=True)
data['Formatted Date'] = pd.to_datetime(data['Formatted Date'],utc = True)
data['month'] = data['Formatted Date'].dt.month
data['year'] = data['Formatted Date'].dt.year
df = data.groupby(['month','year']).mean()
df.head()
df.drop(['index'],axis=1)
```

```python
plt.figure(figsize = (10,5))
a = np.arange(2006,2017)
plt.plot(a,df['Apparent Temperature (C)'][33:44].values, label = 'april Apparent temp.',color = 'navy', marker = 'o')
plt.title('Variation in Apparent temp. in the month of april',fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(loc = 0)
```

```python
plt.figure(figsize = (10,5))
a = np.arange(2006,2017)
plt.plot(a,df['Humidity'][33:44].values, label = 'april Humidity',marker = 'o',color = 'red')
plt.title('Variation in Humidity in the month of april',fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(loc = 0)
```

# variation of temperature in differernt months

```python
plt.figure(figsize=(18,10))
arr= np.arange(2006,2017)
plt.plot(arr,df['Apparent Temperature (C)'][:11].values, label = 'Jan. change in Apparent temp')
plt.plot(arr,df['Apparent Temperature (C)'][11:22].values, label = 'feb. change in Apparent temp.')
plt.plot(arr,df['Apparent Temperature (C)'][22:33].values, label = 'march. change in Apparent temp.')
plt.plot(arr,df['Apparent Temperature (C)'][33:44].values, label = 'april change in Apparent temp.')
plt.plot(arr,df['Apparent Temperature (C)'][44:55].values, label = 'may change in Apparent temp.')
plt.plot(arr,df['Apparent Temperature (C)'][55:66].values, label = 'June change in Apparent temp.')
plt.plot(arr,df['Apparent Temperature (C)'][66:77].values, label = 'July change in Apparent temp.')
plt.plot(arr,df['Apparent Temperature (C)'][77:88].values, label = 'aug. change in Apparent temp.')
plt.plot(arr,df['Apparent Temperature (C)'][88:99].values, label = 'sep. change in Apparent temp.')
plt.plot(arr,df['Apparent Temperature (C)'][99:110].values, label = 'oct. change in Apparent temp.')
plt.plot(arr,df['Apparent Temperature (C)'][110:121].values, label = 'nov. change in Apparent temp.')
plt.plot(arr,df['Apparent Temperature (C)'][121:132].values, label = 'dec. change in Apparent temp.')

plt.legend(loc = 0)

plt.title('Variation in Apparent temperature in diffrent months',fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
```
